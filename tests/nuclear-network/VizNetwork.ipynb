```python
from phloutput import *

import matplotlib.pyplot as plt
```

```python
reactions = [
    r"${^{13}}\mathrm{N} \rightarrow {^{13}}\mathrm{C}$",
    r"${^{14}}\mathrm{O} \rightarrow {^{14}}\mathrm{N}$",
    r"${^{15}}\mathrm{O} \rightarrow {^{15}}\mathrm{N}$",
    r"${^{12}}\mathrm{C} + \mathrm{p} \rightarrow {^{13}}\mathrm{N}$",
    r"${^{13}}\mathrm{C} + \mathrm{p} \rightarrow {^{14}}\mathrm{N}$",
    r"${^{13}}\mathrm{N} + \mathrm{p} \rightarrow {^{14}}\mathrm{O}$",
    r"${^{14}}\mathrm{N} + \mathrm{p} \rightarrow {^{15}}\mathrm{O}$",
    r"${^{15}}\mathrm{N} + \mathrm{p} \rightarrow {^{15}}\mathrm{O} + \alpha$"
]

species = [
    r"p",
    r"${^{4}}\mathrm{He}$",
    r"${^{12}}\mathrm{C}$",
    r"${^{13}}\mathrm{C}$",
    r"${^{13}}\mathrm{N}$",
    r"${^{14}}\mathrm{N}$",
    r"${^{15}}\mathrm{N}$",
    r"${^{14}}\mathrm{O}$",
    r"${^{15}}\mathrm{O}$"
]
```

```python
t, x = timeprof("self.X_species()[:,0,0]")
```

```python
fig, ax = plt.subplots()

ax.plot(t, x)
ax.set_xlabel(r"$t$ (s)")
ax.set_ylabel(r"$X_i$")

ax.set_yscale("log")
ax.set_xscale("log")

ax.legend(species)

plt.show()
```
